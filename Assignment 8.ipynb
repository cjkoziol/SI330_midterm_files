Return a dataframe or series of all mid-cap companies (i.e. market capitalization between $2-10B) along with their total number of employees that started working overseas (those not living in the country where the company is) during the 3rd quarter (Jul-Sep) of 2020. 

```python
company_df = pd.read_csv('EXAM_DATA_COMPANY.csv')
#transact_df = pd.read_csv('EXAM_DATA_TRANSACT.csv')
person_df = pd.read_csv('EXAM_DATA_PERSON.csv')
display(company_df.head(4))
#display(transact_df.head(1))
display(person_df.head(1))
```

```python
company_df = pd.read_csv('EXAM_DATA_COMPANY.csv')
def stuff(x):
    import re
    if pd.isna(x['market_cap']):
        return 
    else:
        result = re.match(r'\$(\d+\.\d+)([BM])', x['market_cap'])
        amount = float(result.group(1))
        multiplier = result.group(2)
        if multiplier == 'M':
            return amount * 1e6
        return amount * 1e9

company_df['market_cap'] = company_df.apply(stuff, axis = 1)
#summary = pd.pivot_table(company_df, values='market_cap', index='country', columns='industry', aggfunc=[np.nanmean]).unstack()
#summary






##mid-cap companies ($2-10B)
newdf = company_df[(company_df['market_cap'] >= 2e9) & (company_df['market_cap'] <= 10e9)]

##total number of employees working overseas
osdf = newdf[newdf['country_code'] != 'US']
mergedf = pd.merge(osdf, person_df, how = 'left', on= 'company') #on = ['country_code', 'address', 'company']

##3rd quarter (jul-Sep) 2020
timedf = mergedf.set_index(pd.to_datetime(mergedf['employee_start'])).sort_index()
finaldf = timedf.loc['2020-07-01 00:00:00':'2020-09-30 23:59:59'].groupby('company').agg({'ssn': len})
finaldf
```
